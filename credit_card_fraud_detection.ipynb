```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/creditcard.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# Checking for null values
data.isnull().sum()
```

```python
# distribution of legit and fraud transaction
data['Class'].value_counts()
```

```python
# Seperating the data since highly unbalanced
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
print(legit.shape, fraud.shape)
```

```python
# Statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare value for both transaction
data.groupby('Class').mean()
```

```python
# Building a new dataset which has a balance
legit_sample = legit.sample(n=492)
```

```python
# Concatenating the dataframe
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# Splitting the data set in x and y
x = new_data.drop(columns='Class', axis=1)
y = new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2, stratify=y)
```

```python
# Training the model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Evaluation of the model
train_predict = model.predict(x_train)
train_accuracy = accuracy_score(train_predict, y_train)
print(train_accuracy)
```

```python
test_predict = model.predict(x_test)
test_accuracy = accuracy_score(test_predict, y_test)
print(test_accuracy)
```
